```
scale_euc_osrm = np.mean(gen_euc_dist_data[0])/np.mean(lam_osrm_dist_data[0])
for i in range(len(gen_euc_dist_data)):
  gen_euc_dist_data[i] = gen_euc_dist_data[i]/scale_euc_osrm
```

```
def r_func(r, D, E):
  return (E <= r).sum() / ((E <= r).sum() + (D > np.mean(D)).sum())
```

```
#D-embedding check 
#r - shepard kruskal - greatest lower bound
D_val_mer = []
D_val_lam = []
r_val_mer =[]
r_val_lam = []
for i in range(len(mer_osrm_dist_data)):
  mer_sqd_err_sum = 0
  euc_sqd_dist_sum = 0
  lam_sqd_err_sum = 0
  for j in range(len(mer_osrm_dist_data[i])):
    mer_sqd_err_sum += (mer_osrm_dist_data[i][j] - gen_euc_dist_data[i][j])**2
    euc_sqd_dist_sum += (gen_euc_dist_data[i][j])**2
    lam_sqd_err_sum += (lam_osrm_dist_data[i][j] - gen_euc_dist_data[i][j])**2
  #mercator r and D
  r_opt_mer = minimize_scalar(lambda r: abs(r_func(r, mer_osrm_dist_data[i], gen_euc_dist_data[i]) - 0.5), bounds=(0, np.max(gen_euc_dist_data[i])), method='bounded').x
  r_val_mer.append(r_opt_mer)
  D_val_mer.append((mer_sqd_err_sum/euc_sqd_dist_sum)*r_opt_mer/300)
  #D_val_mer.append(np.max(np.abs(np.array(mer_osrm_dist_data[i]) - int((np.max(mer_osrm_dist_data[i])/np.max(mer_euc_dist_data[i]))) * np.array(mer_osrm_dist_data[i])))/r_opt_mer)

  #lambert r and D
  r_opt_lam = minimize_scalar(lambda r: abs(r_func(r, lam_osrm_dist_data[i], gen_euc_dist_data[i]) - 0.5), bounds=(0, np.max(gen_euc_dist_data[i])), method='bounded').x
  r_val_lam.append(r_opt_lam)
  D_val_lam.append((lam_sqd_err_sum/euc_sqd_dist_sum)*r_opt_lam/300)
 
x = list(range(len(D_val_mer)))
#print(len(x))
#print(len(D_val_mer))
plt.plot(x, D_val_lam, label = "Distortion Lambert")
plt.axhline(np.mean(D_val_lam), linestyle='--', color='b')
plt.plot(x, D_val_mer, label = "Distortion Mercator")
plt.axhline(np.mean(D_val_mer), linestyle='--', color='orange')
plt.legend()
plt.title("Distortion value between Road and Euclidean distance in D-embedding")
plt.xlabel("Index")
plt.ylabel("Distortion")

plt.show()
```

```
#road network efficiency of each dataset 
osrm_eff = []
for l in lam_osrm_dist_data:
  sum = 0
  cnt = len(l)
  for val in l:
    if val != 0:
      sum += (1/val)
    else:
      cnt = cnt-1
  osrm_eff.append(sum/cnt)
#print(osrm_eff)
euc_eff = []
for l in lam_euc_dist_data:
  sum = 0
  cnt = len(l)
  for val in l:
    if val != 0:
      sum += (1/val)
    else:
      cnt = cnt-1
  euc_eff.append(sum/cnt)
#print(euc_eff)
hav_eff = []
for l in lam_hav_dist_data:
  sum = 0
  cnt = len(l)
  for val in l:
    if val != 0:
      sum += (1/(val))
    else:
      cnt = cnt-1
  hav_eff.append(sum/cnt)

manh_eff = []
for l in lam_manh_dist_data:
  sum = 0
  cnt = len(l)
  for val in l:
    if val != 0:
      sum += (1/(val))
    else:
      cnt = cnt-1
  manh_eff.append(sum/cnt)
```

```
def f_static(D1, D2):
    n = len(D1)
    F = np.sum((D1 - D2)**2) / ((n*(n-1)/2) * np.var(np.hstack((D1, D2))))
    return F
```

```
#F-static
F_hav_list = []
F_euc_list = []
F_manh_list = []
for i in range(len(lam_euc_dist_data)):
  F_hav = f_static(np.array(lam_osrm_dist_data[i]), np.array(lam_hav_dist_data[i]))
  F_euc = f_static(np.array(lam_osrm_dist_data[i]), np.array(lam_euc_dist_data[i]))
  F_manh = f_static(np.array(lam_osrm_dist_data[i]), np.array(lam_manh_dist_data[i]))
  F_hav_list.append(F_hav)
  F_euc_list.append(F_euc)
  F_manh_list.append(F_manh)

x = list(range(len(F_hav_list)))
#print(len(x))
#print(len(F_hav_list))
plt.plot(x, F_hav_list, label = "F-static with Haversine", linestyle='--', color = 'b')
plt.axhline(np.mean(F_hav_list), color='b')
plt.plot(x, F_euc_list, label = "F-static with Euclidean", linestyle='--', color = 'orange')
plt.axhline(np.mean(F_euc_list), color='orange')
plt.plot(x, F_manh_list, label = "F-static with Manhattan", linestyle='--', color = 'green')
plt.axhline(np.mean(F_manh_list), color='green')
# plt.plot(x, hav_eff, label = "Haversine eff", linestyle='--')
# plt.axhline(np.mean(hav_eff), color='green')
plt.legend()
plt.title("F-static between Road and Euclidean, Haversine, Manhattan distance")
plt.xlabel("Index")
plt.ylabel("F-static")

plt.show()
```
